# Proyecto de regresion lineal


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import balanced_accuracy_score, roc_auc_score, make_scorer
from sklearn.model_selection import train_test_split, GridSearchCV # cross validation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
import graphviz

# ´from sklearn.metrics import plot_confusion_matrix # draws a confusion matrix
# aparentemente esta deprecada la funcion plot_confusion_matrix  de sklearn
import seaborn as sns

df = pd.read_excel("/Users/tomasgarcia/Downloads/Telco_customer_churn.xlsx")

print(f'...........................{df.columns}')

""" vamos a eliminar la columna churn reason porque es algo que los usuarios 
completan cuando ya no quieren usar mas el servicio. entonces no nos servira eso
para predecir"""
df.drop(["Churn Reason"], axis=1, inplace=True) # sin el inplace=True no lo
# hara en el dataframe original

df.drop(['CustomerID', 'Churn Label', 'Churn Score', 'CLTV'],
axis=1, inplace=True) # ojo con pasar alguna columna que no esta porque
# arrojara error key error. por ejemplo,
# KeyError: "['Churn Reason'] not found in axis"

''' 
sacaremos todas las que tienen un solo valor, 
pues no nos ayudan a predecir
'''
print(df['Count'].unique())
print((df['Country'].unique()))
print((df['State'].unique()))
print((df['City'].unique()))

# Lista para almacenar las columnas a eliminar
columnas_a_eliminar = []

# Iteramos sobre todas las columnas del DataFrame
for columna in df.columns:
    # Si la cantidad de valores únicos en la columna es igual a 1, significa que tiene un solo valor
    if len(df[columna].unique()) == 1:
        # Agregamos la columna a la lista de columnas a eliminar
        columnas_a_eliminar.append(columna)

# Eliminamos las columnas del DataFrame
df = df.drop(columns=columnas_a_eliminar)

# Mostramos el DataFrame resultante
print("DataFrame sin las columnas con un solo valor:")
print(df)

df.replace({'City': {' ': '_'}}, regex=True, inplace=True)
df.columns = df.columns.str.replace(" ", "_")

df.loc[(df ['Total_Charges'] == ' '), 'Total_Charges'] = 0

df['Total_Charges'] = pd.to_numeric(df['Total_Charges'])
print(df.dtypes)

df.replace(" ", "_", regex=True, inplace=True)

# recordemos que esto de rellenar los espacios en blanco con un _ es para
# poder graficarlo luego. En cuanto al XGBoost no hay diferencia aca

X = df.drop('Churn_Value', axis=1).copy() # alternatively: X = df_no_missing iloc[:,:-1]
X.head()
# el .copy es para no borrar eso en la columna (como ocurriria si aplicaramos el inplace=True) sino en una copia

y = df['Churn_Value'].copy()
print(y.head())

X_encoded = pd.get_dummies(X, columns=['City',
                                       'Gender',
                                       'Senior_Citizen',
                                       'Partner',
                                       'Dependents',
                                       'Phone_Service',
                                       'Multiple_Lines',
                                       'Online_Security',
                                       'Online_Backup',
                                       'Device_Protection',
                                       'Tech_Support',
                                       'Streaming_TV',
                                       'Streaming_Movies',
                                       'Contract',
                                       'Paperless_Billing',
                                       'Payment_Method',
                                       'Lat_Long',
                                       'Internet_Service'
                                       ])

print(X_encoded.head())


X_train, X_test, Y_train, Y_test = train_test_split(X_encoded,
                                                    y, random_state=42,
                                                    stratify=y)
""" LA ESTRATIFICACION
al hacer esto se mantienen las proporciones en cada grupo. 
Es decir, en el train_set habra un 27% de una categoria y 
un 73% de la otra, y lo mismo ocurre en el test_Set."""

print(f'...........................{df.columns}')

# construimos el clasificador

clf_xgb = xgb.XGBClassifier(objective='binary:logistic', missing=np.nan, seed=42)
clf_xgb.fit(X_train,
            Y_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_metric='aucpr',
            eval_set=[(X_test, Y_test)])

# eso de missing=None no hace falta, ya que el default es None
# la seed hace referencia a la aleatoriedad
# el verbose es para que se muestren mensajes durante el proceso
# el early_stopping_rounds=10 hace que cuando no se este mejorando siga 10 vueltas mas
# el eval_metric='aucpr' lo que hace es ir evaluando el rendimiento del modelo en el test set

'''Eso del eval_set lo que hace es hacer el cross_validation, 
en lugar de tener que hacerlo nosotros a mano el XGBoost lo hace por nosotros.'''

# Ahora creamos una confusion matrix para ver como funciona
Y_pred = clf_xgb.predict(X_test)

# Calcular la matriz de confusión
cm = confusion_matrix(Y_test, Y_pred)

# Visualizar la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf_xgb.classes_)
disp.plot()
# plt.show()

""" eso nos muestra que cuando las personas dejan la empresa no se las clasifica bien, masomenos un 50 de error. esto se da principalmente porque
los datos estan desbalanceados. entonces lo que nos conviene hacer es implementar un GridSearchCv, lo cual luego se vera bien porque tiene sus 
complicaciones, pero basicamente es algo para encontrar la mejor combinacion de hiperparametros. asi que vamos a suponer que ya hicimos eso y 
vamos a poner el modelo ahora con esos hiperparametros encontrados:"""

clf_xgb = xgb.XGBClassifier(seed=42,
                            objective='binary:logistic',
                            gamma=0.25,
                            Learn_rate=0.1,
                            max_depth=4,
                            reg_lambda=10,
                            scale_pos_weight=3,
                            subsample=0.9,
                            colsample_bytree=0.5)

clf_xgb.fit(X_train,
            Y_train,
            verbose=True,
            early_stopping_rounds=10,
            eval_metric='aucpr',
            eval_set=[(X_test, Y_test)])

# ahora grafiquemos la confusion matrix para ver si mejoro
Y_pred = clf_xgb.predict(X_test)

# Calcular la matriz de confusión
cm = confusion_matrix(Y_test, Y_pred)

# Visualizar la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=clf_xgb.classes_)
disp.plot()
plt.show(block=False)

# ahora vamos a graficar un arbol (aca hay varios arboles [al igual que en bagged o en rf], pero vamos a graficar uno para entender las cosas)


bst = clf_xgb.get_booster()
for importance_type in ('weight', 'gain', 'cover', 'total_gain', 'total_cover'):
    print('%s: ' % importance_type, bst.get_score(importance_type=importance_type))
node_params = {'shape': 'box', ## make the nodes fancy
                'style': 'filled, rounded',
                'fillcolor': '#78cbe'}
leaf_params= {'shape': 'box',
             'style': 'filled',
            'fillcolor': '#e48038'}
## NOTE: num_trees is NOT the number of trees to plot, but the specific tree you want to plot ## The default value is 0, but I'm setting it just to show it in action since it is ## counter-intuitive.
xgb.to_graphviz(clf_xgb,num_trees=0, size="10,10",
condition_node_params=node_params,
leaf_node_params=leaf_params)
## if you want to save the figure...
# graph_data = xgb.to_graphviz(clf_xgb, num_trees=0, size="10,10",
#condition_node_params=node_params, leaf_node_params=leaf_params)
# graph_data.view(filename='xgboost_tree_customer_churn') ## save as PDF
```
